Match ISINs and identify the column(s) containing them 
REgex ([A-Z]{2})([A-Z0-9]{9})([0-9]{1})  2 alpha followed by 9 numeric folled by check digit

```python
import re
import numpy
from itertools import repeat

# s1.str.contains('PARROT', flags=re.IGNORECASE, regex=True)


ISINRegex = re.compile(r'([A-Z]{2})([A-Z0-9]{9})([0-9]{1})',  re.IGNORECASE)
dateRegex = re.compile(r'([A-Z]{2})([A-Z0-9]{9})([0-9]{1})',  re.IGNORECASE)
GICSRegex = re.compile(r'([A-Z]{2})([A-Z0-9]{9})([0-9]{1})',  re.IGNORECASE)
NAICSRegex = re.compile(r'([A-Z]{2})([A-Z0-9]{9})([0-9]{1})',  re.IGNORECASE)
currencyREGEX = re.compile(r'([A-Z]{2})([A-Z0-9]{9})([0-9]{1})',  re.IGNORECASE)
NAICSREGEX = re.compile(r'([A-Z]{2})([A-Z0-9]{9})([0-9]{1})',  re.IGNORECASE)

recordCount = df.shape[0]
columnCount = len(df.columns)
print("The Total number of Rows "+str(recordCount ))
print("The Total number of Columns "+str(columnCount) )

ISINCount = 0
GICSCount = 0
NAICSCount = 0
dateCount=0
currencyCount=0
uomCount=0

# Convert all values to string
df = df.astype('string')
# print('ISIN found: '+"% s" % len(ISINMatches))
# metricsCount = 5
# results = [[] for x in repeat(None, columnCount)]
# print(results)
results = []
resultsHeader = ['COLUMN', 'ISIN MATCH COUNT', 'GICS MATCH COUNT', 'NAICS MATCH COUNT', 'DATE MATCH COUNT', 'CURRENCY MATCH COUNT','UOM COUNT', 'RECORD COUNT']

for (columnName, columnData) in df.iteritems():
    # print(columnName)
    dataList = list(columnData.values)
    S=pd.Series(dataList)

    # S=pd.Series(['CX_IE00BY7QL619_NYQ', 'US0846707026', 'US0846707026'])
    # if (columnName == 'UNIQUEID') :
    # print(ISINCount)
   
    ISINCount = S.str.count(r'([A-Z]{2})([A-Z0-9]{9})([0-9]{1})').sum()


    print("ISINCount "+str(ISINCount))
    # print("ISINCount "+str(ISINCount))
    # ISINMatches = ISINRegex.str.findall('LR0008862868 US6935061076US00206R1023 CX_US70450Y1038_NSQ')
    # ISINCount = len(ISINMatches)
    results.append([columnName,ISINCount,  NAICSCount, GICSCount, dateCount,currencyCount, uomCount, recordCount])  
 

# print(results)
results_df = pd.DataFrame(results, columns =resultsHeader)
results_df
```
